```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline

# Load the data again (or ensure 'df' is the original DataFrame)
df=pd.read_csv('/content/Customer_support_data.csv')
df.head()

# Drop rows with missing values in 'Item_price' or 'CSAT Score'
df_cleaned = df.dropna(subset=['Item_price', 'CSAT Score']).copy() # Use .copy()

# Scale the data after dropping missing values
scalar = MinMaxScaler()
df_cleaned['Item_price'] = scalar.fit_transform(df_cleaned[['Item_price']])
df_cleaned['CSAT Score'] = scalar.fit_transform(df_cleaned[['CSAT Score']])

# Apply KMeans on the cleaned and scaled data
km = KMeans(n_clusters=3, random_state=42) # Added random_state for reproducibility
y_predicted = km.fit_predict(df_cleaned[['Item_price', 'CSAT Score']])

# ***Assign the cluster predictions to the cleaned DataFrame here***
df_cleaned['cluster'] = y_predicted

# Merge the cluster information back to the original DataFrame
# We use the index to align the data.
df = df.merge(df_cleaned[['cluster']], left_index=True, right_index=True, how='left')

# Now the original df has the 'cluster' column with cluster IDs for rows
# that were included in clustering, and NaN for rows that were dropped.

# You can now proceed with analyzing or visualizing the clusters in the original df
print(df.head())

# Plotting the clusters (using the original df with the new 'cluster' column)
# Handle potential NaNs in 'cluster' if they exist after the merge
df1 = df[df.cluster==0.0] # Note: cluster column might be float due to merge
df2 = df[df.cluster==1.0]
df3 = df[df.cluster==2.0]
# You might also want to handle rows where cluster is NaN if you didn't fill them

plt.scatter(df1.Item_price,df1['CSAT Score'],color='green',marker='+')
plt.scatter(df2.Item_price,df2['CSAT Score'],color='red',marker='.')
plt.scatter(df3.Item_price,df3['CSAT Score'],color='blue',marker='*')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centeroid')
plt.legend()
plt.show() # Add plt.show() to display the plot
```
